```python
import random
import matplotlib.pyplot as plt
import numpy as np
from sklearn import svm
```

```python

random.seed(314)


x_vals = [random.random() for _ in range(40)]
y_vals = [random.random() for _ in range(40)]


class_labels = [(2 * x < y + 0.5) + 1 for x, y in zip(x_vals, y_vals)]


class1_data = [[x, y] for x, y, label in zip(x_vals, y_vals, class_labels) if label == 1]
class2_data = [[x, y] for x, y, label in zip(x_vals, y_vals, class_labels) if label == 2]


class1_x = [data[0] for data in class1_data]
class1_y = [data[1] for data in class1_data]
class2_x = [data[0] for data in class2_data]
class2_y = [data[1] for data in class2_data]

plt.scatter(class1_x, class1_y, c='blue', label='Class 1')
plt.scatter(class2_x, class2_y, c='red', label='Class 2')
```

```python
features = [[x, y] for x, y in zip(x_vals, y_vals)]

svm_classifier = svm.SVC(kernel='linear', C = 1e200)
svm_classifier.fit(features, class_labels)

weights = svm_classifier.coef_[0]
slope = -weights[0] / weights[1]
xx = np.arange(0,1,0.01)
yy = [slope * x - (svm_classifier.intercept_[0]) / weights[1] for x in xx]

```

```python
plt.scatter(class1_x, class1_y, c='blue', label='Class 1')
plt.scatter(class2_x, class2_y, c='red', label='Class 2')
plt.plot(xx, yy, 'k-', label='Decision Boundary')
plt.title("SVM Decision Boundary")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
